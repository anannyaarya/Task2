**1. Import Libraries & Load Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

**2. Generate Summary Statistics**

```python
df.describe()
df.info()
df.isnull().sum()
```

**3. Visualizations**

3.1 Histograms for Numeric Features

```python
df.hist(figsize=(10, 8))
plt.suptitle('Histograms of Numeric Features')
plt.show()
```

3.2 Boxplots for Outlier Detection

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplot for Age and Fare")
plt.show()
```

3.3 Pairplot for Feature Relationships

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']], hue='Survived')
```

3.4 Correlation Matrix & Heatmap

```python
correlation = df.select_dtypes(include=np.number).corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

3.5 Plotly Interactive Visualization

```python
fig = px.histogram(df, x='Age', color='Survived', barmode='overlay')
fig.show()
```
